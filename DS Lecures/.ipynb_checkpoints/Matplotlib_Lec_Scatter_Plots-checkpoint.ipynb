```python
import matplotlib.pyplot as plt, pandas as pd, numpy as np
%matplotlib inline
```

```python
x = np.random.randint(1,9,20)
y = np.random.randint(1,9,20)
```

```python
print(x, y)
```

```python
sat = np.random.randint(1,5,20)
sizes = np.random.randint(150,550,20)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x,y, s= 200, c = 'green', marker = 'o', edgecolors= 'k', 
            linewidths=2, alpha = 0.5
)
```

```python
plt.figure(figsize = (10,6))
plt.scatter(x,y, s= sizes, c = sat, marker = 'o', edgecolors= 'k', 
            linewidths=2, alpha = 0.75, cmap = 'Greens'
)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')
plt.show()

```

```python
data = pd.read_csv(r"C:\Users\sunny\Desktop\sample_data\like_views.csv")
data.head()
```

```python
view_counts = data.view_count
likes = data.likes
ratio = data.ratio
```

```python
plt.scatter(view_counts, likes, edgecolor = 'black',
            c = ratio, cmap = 'summer' ,linewidth = 1, alpha = 0.75)

cbar = plt.colorbar()
cbar.set_label('Like/Dislike ratio')

plt.title('Trending Youtube Videos')
plt.xlabel('View Conts')
plt.ylabel('Total Lime')

# plt.xscale('log')
# plt.yscale('log')

plt.tight_layout()
plt.show()
```
