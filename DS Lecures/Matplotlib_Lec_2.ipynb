```python
import matplotlib.pyplot as plt, numpy as np, pandas as pd
%matplotlib inline
```

## Here is some additional functionality for line graphs

```python
plt.figure(figsize = (10,4))
x = np.arange(1, 10)**2
y = np.arange(1, 20,3)**2
z = np.arange(1, 20,4)**2

plt.plot(x, label = 'X', lw = 5, marker = '^', markersize = 10, 
        color = '#87d4ed', linestyle = '-.')
plt.plot(y, label = 'Y', lw = 5, marker = 'd', markersize = 10,
        color = '#87edd0', linestyle = '--')
plt.plot(z, label = 'Z', lw = 5, marker = 'o', markersize = 10, 
        color = '#e6ed87', linestyle = ':')

plt.xticks(fontsize = 15, color = '#d97e99') ; plt.yticks(fontsize = 15, color = '#7ed9c5')
plt.xlabel('Time', fontsize = 20, color = '#ffe224')
plt.ylabel('Growth', fontsize = 20, color = '#ff4c24')

plt.title('Time Vs Growth Graph', fontsize = 30, color = '#e00d7e', loc = 'center')

plt.legend(fontsize = 15, loc =  (0.1,0.5))

plt.grid(True)
```

```python

data = [np.random.randint(1,20,10),
np.random.randint(1,20,10),
np.random.randint(1,20,10)]

for i in range(len(data)):
    plt.plot(data[i], label = ['A', 'B', 'C'][i], marker = ['^','d','o'][i], 
            color = ['pink', 'b','g'][i], markersize = 15, lw = 5,
            markerfacecolor = ['g', 'r', 'yellow'][i], linestyle = ['-','--','-.'][i], 
            markeredgewidth = 5)
    plt.grid(True)
    plt.show()
```

# With List Comprehension 

```python

data = [np.random.randint(1,20,10),
np.random.randint(1,20,10),
np.random.randint(1,20,10)]
title=['graph1','graph2','graph3']
[(
    plt.plot(data[i], label = ['A', 'B', 'C'][i], marker = ['^','d','o'][i], 
            color = ['pink', 'b','g'][i], markersize = 15, lw = 5,
            markerfacecolor = ['g', 'r', 'yellow'][i], linestyle = ['-','--','-.'][i], 
            markeredgewidth = 5),
    plt.title(title[i], fontsize = 30, color = '#e00d7e', loc = 'center'),
    plt.grid(True),
    plt.show()
    )for i in range(len(data))]
```
