# Load File Schools in Punjab

```python
columns=['District', 'Male', 'Female', 'H.S Total', 'Male', 'Female', 'H Total', 'Male', 'Female', 'Midl Total', 'Male', 'Female', 'P Total','Male','Female','T Mosq','Male Total','Female Total','Grand Total']
df2 = pd.read_excel('SchoolsinPunjab.xlsx')
df2.head()
```

```python
df3 = df2.loc[:,['District','HS_Total','High_Total','Midl_Total','Prim_Total','Mosq_Total',]]
df3.set_index('District' , inplace = True)
df3.head(38)
```

```python
df.set_index(['Date','District'], inplace = True)
df.head()
```

```python
df['Prim_Ratio'] = df['Total Primary']/df3['Prim_Total']
df['Midl_Ratio'] = df['Total Middel']/df3['Midl_Total']
df['High_Ratio'] = df['Total High']/df3['High_Total']
df['HS_Ratio']   = df['Total High Secondary']/df3['HS_Total']
#df['Mosq_Ratio'] = df['Total Primary']/df3['Prim_Total']
df.head(10)
```

```python
df.reset_index(inplace = True)
df3.reset_index(inplace = True)
```

```python
plt.figure(figsize = (10,4))
plt.xlabel('Date')

plt.ylabel('Ratio(Number of student in school)')

plt.plot(df.Date, df.Prim_Ratio,label ='Primary School Ratio')

#plt.plot(df.Date, df.Midl_Ratio,  label ='Middle Schoo Ratio')

#plt.plot(df.Date, df.High_Ratio, label='High School Ratio')

plt.legend()

# To add some padding we can run a method
plt.tight_layout()
```

## Primary Schools 

```python
x = df3.District
y = df3.Prim_Total
sizes = df.Prim_Ratio

plt.figure(figsize = (15,6))
plt.scatter(x,y, s= sizes,  marker = 'o', edgecolors= 'k', 
            linewidths=2, alpha = 0.75, cmap = 'Greens')
plt.xticks(rotation=90)

#cbar = plt.colorbar()
#cbar.set_label('Satisfaction')
plt.xlabel('District')
plt.ylabel('Ratio(Number of student in Primary School)')
plt.show()
```
