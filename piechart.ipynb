```python
import matplotlib.pyplot as plt, numpy as np, pandas as pd 
%matplotlib inline
```

```python
plt.style.use('Solarize_Light2')
slices = [60,40, 25, 35]
labels = ['Boys', 'Girls', 'Kids', 'Adults']
colors = ['red', 'blue', 'yellow','green']
plt.pie(slices, labels = labels, wedgeprops= {'edgecolor' : 'white'} )
plt.show()
```

```python
explode = [0,0,0,0.15,0]
languages = [ 'Java', 'HTML/CSS', 'SQL','Python' , 'JavaScript']
popularity = [35917,36443,47544,55466,59219]
```

```python
plt.pie(popularity, labels = languages, explode = explode, rotatelabels = False, 
wedgeprops = {'edgecolor' : 'black'}, shadow = True, startangle= 90, autopct= '%1.1f%%',
       frame= False, labeldistance=1.1)
plt.show()
```
